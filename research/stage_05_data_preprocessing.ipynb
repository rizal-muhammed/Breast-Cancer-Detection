```python
input_dir = Path("artifacts/training_file_from_db")
```

```python
def load_breast_cancer_wisconsin_data(breast_cancer_wisconsin_path=input_dir):
    csv_path = os.path.join(breast_cancer_wisconsin_path, "inputfile.csv")
    return pd.read_csv(csv_path, index_col='index_column')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Replace 'target_feature' with the actual name of your target feature
target_feature = 'target_feature'

# Sample DataFrame creation
data = {'feature1': [1, 2, 3, 4, 5],
        'feature2': [5, 4, 3, 2, 1],
        'target_feature': [0, 1, 1, 0, 1]}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix[[target_feature]], annot=True, cmap='coolwarm', fmt=".2f")
plt.title(f'Correlation Matrix with Respect to {target_feature}')
plt.show()
```

```python
bcw['diagnosis'] = bcw['diagnosis'].map({'M': 1, 'B': 0})
corr_matrix = bcw.corr()

corr_matrix['diagnosis'].sort_values(ascending=False)
```

```python
target_feature = 'diagnosis'
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix[[target_feature]].abs().sort_values(by=target_feature, ascending=False), annot=True, cmap='coolwarm', fmt=".2f")
plt.title(f'Correlation Matrix with Respect to {target_feature}')
plt.show()
```
